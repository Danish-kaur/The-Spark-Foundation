```python
plt.figure(figsize=(10,10))
data['Sub-Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
sns.countplot(x=data['State'],order=(data['State'].value_counts().head(20)).index)
plt.xticks(rotation=90)
```

```python
sns.countplot(x =data['State'], order=(data['State'].value_counts().tail(33)).index)
plt.xticks(rotation=90)
```

```python
states = data.groupby('State')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending = False)
plt.figure(figsize=(30,30))
states[:25].plot(kind = 'bar', edgecolor="#000000", color=['yellow', 'blue'])
plt.title('Status of top 30 States')
plt.xlabel('Sates')
plt.ylabel('Total')
plt.grid(True)
states[25:].plot(kind = 'bar', edgecolor="#000000", color=['yellow','blue'])
plt.title("Status of bottom 30 States")
plt.xlabel('States')
plt.ylabel("Total")
plt.grid(True)
```

```python
data.groupby('Sub-Category')['Profit', 'Sales'].sum().plot.bar(color=['yellow', 'blue'])
plt.figure(figsize = (40,40))
```

```python
pd.DataFrame(data.groupby('State').sum())['Profit'].sort_values(ascending=True)
pd.DataFrame(data.groupby('State').sum())['Discount'].sort_values(ascending=True)
```

```python
plt.figure(figsize = (29,9))
sns.lineplot('Discount', 'Profit', data=data, color = 'yellow', label='Discount')
plt.legend()
```
